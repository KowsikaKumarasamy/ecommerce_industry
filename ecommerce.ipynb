```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## renaming the column name

df.rename(columns = {'category_1': 'overview',
                    'category_2': 'category',
                    'category_3': 'specific'},inplace = True)
df
```

```python
df.drop(['description','highlights','image_links'],axis = 1,inplace = True)
df
```

```python
df.info()
```

```python
df['selling_price']= df['selling_price'].fillna(0).replace('₹','',regex = True).replace(',','',regex = True).astype(int)
df['mrp'] = df['mrp'].fillna(0).replace('₹','',regex = True).replace(',','',regex = True).astype(int)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isna().sum()
```

```python
df[df['title'].isna()]
```

```python
df.shape
```

```python
df = df.dropna(subset=['title'])
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[df['seller_name'].isna()]
```

```python
###assigning values to the NaN fields

df['product_rating'] = df['product_rating'].fillna(df['product_rating'].mode()[0])
df['seller_name'] = df['seller_name'].fillna('Unknown')
df['seller_rating'] = df['seller_rating'].fillna(df['seller_rating'].mode()[0])
df.isna().sum()
```

```python
### visualization
```

```python
# Overview of categories using pie

Overview_counts = df['overview'].value_counts()
Overview_counts.plot(kind = 'pie', 
                     autopct = '%1.1f%%', 
                     startangle = 90, 
                     figsize = (5,5), colors = sns.color_palette('Set2'),
                     pctdistance = 0.80)
plt.title('Category overview')
plt.ylabel('')
hole = plt.Circle((0,0),0.65, facecolor = 'White')
plt.gcf().gca().add_artist(hole)
```

```python
#Top 10 products

vc = df['specific'].value_counts()[0:10]

plt.figure(figsize = (15,5))
g = sns.barplot(x=vc.index, y = vc.values, palette = 'Set2')

for i in range(10):
    value = vc[i]
    g.text(y = value-2, x = i+0.125, s = value, color = 'black',ha = 'center', fontsize = 10)

plt.title('Top 10 Products', fontsize = 20, color = 'red')
plt.show()
```

```python
#Top 10 Seller name

vc = df['seller_name'].value_counts()[0:10]

plt.figure(figsize = (20,5))
g = sns.barplot(x=vc.index, y = vc.values, palette = 'Set2')

for i in range(10):
    value = vc[i]
    g.text(y = value-2, x = i+0.125, s = value, color = 'black',ha = 'center', fontsize = 10)

plt.title('Top 10 Sellers', fontsize = 20, color = 'red')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# First plot
plt.subplot(2, 1, 1)  
sns.boxplot(data=df, x='selling_price', y='overview', palette='Set1')
plt.title('Product by Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Product Overview')

# Second plot
plt.subplot(2, 1, 2) 
sns.boxplot(data=df, x='mrp', y='overview', palette='Set2')
plt.title('Product by MRP')
plt.xlabel('MRP')
plt.ylabel('Product Overview')

plt.tight_layout()


plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='overview', y='selling_price', data=df, palette='Set3')

# Adding titles and labels
plt.title('Selling Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Selling Price')

plt.show()
```
